```python
# Lendo o CSV contendo os dados
dados = pd.read_csv("dados.csv").drop(columns=['#','Nível'])
dados = dados[dados["Município"].str.endswith("(RJ)")]

pd.set_option('display.max_rows', len(dados))
dados.style.hide(axis="index")

análise = dados[['2007']].rename(columns={"2007" : "População"})

dados
```
