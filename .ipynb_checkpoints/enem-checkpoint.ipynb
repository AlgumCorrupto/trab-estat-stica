# Análise das notas da prova de matemática do Enem 2024

Análise básica sobre as notas dos alunos que fizeram a prova no município de Campos dos Goytacazes na prova de matemática do Enem 2024 

## Passo 0: importando as dependências

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

## Passo 1: Preparando os dados para a análise

```python
dados_brasil = pd.read_csv("dados.csv", sep=';', encoding='latin1')
população_campos = dados_brasil[dados_brasil["NO_MUNICIPIO_ESC"].str.lower() == "campos dos goytacazes"][['NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]

amostra = população_campos.dropna().sort_values(by='NU_NOTA_MT')
notas = np.round(np.array(amostra["NU_NOTA_MT"], dtype=np.float32), decimals=0, out=None)

print(f"{len(amostra)} participantes fizeram a prova de matemática")
amostra
```

## Passo 2 - Calculando medidas centrais

```python
média = np.mean(notas)
mediana = np.median(notas)
moda = st.mode(notas).mode

print(f"Média: {média}", f"Mediana: {mediana}", f"Moda: {moda}", sep="\n")
```

## Passo 3 - Calculando medidas de dispersão

```python
amplitude = notas[-1] - notas[0]
variância = sum([np.pow(xi - média, 2) for xi in notas]) / len(notas)
desvio_padrão = np.sqrt(variância)
cv = desvio_padrão / média
#pcc 

print(f"Amplitude: {amplitude}", f"Desvio padrão: {desvio_padrão}", f"Coeficiente de variação: {cv}", sep='\n')
```

## Passo 4 - Calculando medidas de assimetria

```python
assimetria = st.skew(notas)
curtose = st.kurtosis(notas)
pdf = st.norm.pdf(notas, média, desvio_padrão)

print(f"Assimetria: {assimetria}", f"Curtose: {curtose}", sep='\n')
```

## Passo 5 - Visualizando os dados

```python
# histograma das provas do enem

# pegando notas de outras categorias para comparação
# ciências naturais
notas_cn = [i for i in np.round(np.array(amostra["NU_NOTA_CN"], dtype=np.float32), decimals=0, out=None) if i != 0]

# redação
notas_re = [i for i in np.round(np.array(amostra["NU_NOTA_REDACAO"], dtype=np.float32), decimals=0, out=None ) if i != 0]

# Você pode comentar uma das linhas para ver o histograma de cada tipo mais claramente
hist_re = plt.hist(notas_re, bins=55, density=True, alpha=0.75, color='#091832', label="Redação")
qtd_re, edge_re = hist_re[0], hist_re[1]

hist_cn = plt.hist(notas_cn, bins=55, density=True, alpha=0.75, color='#7d426c', label="Ciências Naturais")
qtd_cn, edge_cn = hist_cn[0], hist_cn[1]

hist_mt = plt.hist(notas, bins=55, density=True, alpha=0.75, color='#ffc281', label="Matemática")
qtd, edge = hist_mt[0], hist_mt[1]

plt.plot(notas, pdf, color="#00BBF9", label="Pdf da matemática")
plt.legend()
plt.show()
```

```python
# Polígono de frequência

mid = (edge[:-1] + edge[1:]) / 2
mid_re = (edge_re[:-1] + edge_re[1:]) / 2
mid_cn = (edge_cn[:-1] + edge_cn[1:]) / 2

plt.plot(mid_re, qtd_re, marker='o', linestyle='-', color='#091832', label='Redação')
plt.plot(mid_cn, qtd_cn, marker='o', linestyle='-', color='#7d426c', label='Ciências Naturais')
plt.plot(mid, qtd, marker='o', linestyle='-', color='#ffc281', label='Matemática')

plt.legend()
plt.show()
```
